Fonction qui permet d'encoder 

```python
def encode_target(df, target_column):
    """Add column to df with integers for the target.

    Args
    ----
    df -- pandas DataFrame.
    target_column -- column to map to int, producing
                     new Target column.

    Returns
    -------
    df_mod -- modified DataFrame.
    targets -- list of target names.
    """
    df_mod = df.copy()
    targets = df_mod[target_column].unique()
    map_to_int = {name: n for n, name in enumerate(targets)}
    df_mod["Operation"] = df_mod[target_column].replace(map_to_int)

    return (df_mod, targets)
```

```python
y_train=trainData.iloc[:,8].copy() 
x=trainData.iloc[:,:8].copy()

df1, targets = encode_target(x, "Type Operation")
print(targets)
```

```python
features=['Montant', 'Montant Cpt Source avt', 'Montant Cpt Source aft', 'Montant Cpt Dest Avt', 'Montant Cpt Dest Aft', 'Operation']
X_train = df1[features]
```

```python
tree.fit(X_train, y_train)
```

```python
y_test=testData.iloc[:,8].copy() 
x_test=testData.iloc[:,:8].copy() 

df2, targets = encode_target(x_test, "Type Operation")

X_test = df2[features]

```

```python

##training_accuracy = clf.score(X, y)
test_accuracy = tree.score(X_test, y_test)
train_accuracy = tree.score(X_train, y_train)
##print("Training accuracy:", training_accuracy)
print("Train accuracy:", train_accuracy)
print("Test accuracy:", test_accuracy)
#On obtient une précision très faible
```
